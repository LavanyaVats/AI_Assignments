1. Create and Inspect an Array Create a 1D NumPy array containing integers from 10 to 50 (inclusive) with a step of 5. Print its shape, size, and data type.

```python
arr1 = np.arange(10,51,5)
print("1. Array:", arr1)
print("Shape:", arr1.shape)
print("Size:",arr1.size)
print("Data Type :",arr1.dtype)
```

2. Generate a Matrix with Random Values
    Generate a 4x5 NumPy array with random integers between 1 and 100.

    Find the maximum and minimum values, along with their indices.
    


```python
arr2 = np.random.randint(1,100,(4,5))
print("\n2. Random Matrix: \n", arr2)
print("Maximum:", arr2.max(), "at index", np.unravel_index(arr2.argmax(), arr2.shape))
print("Minimum:", arr2.min(), "at index", np.unravel_index(arr2.argmax(), arr2.shape))
```

3. Reshape an Array
    Create a 1D array of numbers from 1 to 24, then reshape it into a 4x6 matrix.

    Extract the 2nd row and 3rd column.

```python
arr3 = np.arange(1,25).reshape(4,6)
print("\n3. Reshaped Array:\n", arr3)
print("2nd Row : ", arr3[1, :])
print("3rd Column : ", arr3[:, 2])
```

4. Boolean Indexing
    Given a NumPy array of 15 random integers between 0 and 50, find all elements greater than 25 and replace them with -1.

```python
arr4 =np.random.randint(0,51,15)
print("Original Array : ", arr4)
arr4[arr4 > 25] = -1
print("Modified array : ", arr4)
```

5. Mathematical Operations
    Create two 1D NumPy arrays:
    a = [1, 2, 3, 4, 5] and b = [10, 20, 30, 40, 50]

    Perform element-wise addition, subtraction, multiplication, and division.


```python
a = np.array([1,2,3,4,5])
b = np.array([10,20,30,40,50])
print("\5. Addition : ", a+b)
print("Subtraction : ", a-b)
print("Multiplication: ", a*b)
print("Division: ", a/b)
```

6. Statistical Computations
    Generate a 5x5 matrix with random floats between 0 and 1.

    Compute the mean, median, variance, and standard deviation.

```python
arr6 = np.random.rand(5,5)
print("\n6. Random 5x5 Matrix:\n", arr6)
print("Mean:", arr6.mean())
print("Median:", np.median(arr6))
print("Variance:", arr6.var())
print("Standard Deviation:", arr6.std())
```

7. Matrix Multiplication
    Create two matrices:

    A is 3x2 with values from 1 to 6.

    B is 2x3 with values from 7 to 12.
    Perform matrix multiplication (A @ B) and display the result.

```python
A = np.arange(1,7).reshape(3,2)
B = np.arange(7,13).reshape(2,3)
print("\n7. Matrix A:\n", A)
print("Matrix B:\n", B)
print("\n7. Matrix Multiplication:\n", A@B)
```

8. Slice and Modify
    Create a 6x6 array of integers from 1 to 36.

    Extract the 3x3 sub-matrix from the center and set all its values to 0.

```python
arr8 = np.arange(1,37).reshape(6,6)
print("\n8. Original 6x6 Matrix:\n", arr8)
arr8[2:5, 2:5] = 0
print("Modified Matrix with center 3x3 set to 0:\n", arr8)
```

9. Use of np.where
    Create a NumPy array with values [5, 10, 15, 20, 25].

    Replace all elements greater than 15 with 99, otherwise replace with 0.

```python
arr9 = np.array([5,10,25,20,25])
arr9_result = np.where(arr9 >15, 99, 0)
print("\n9. np.where Result:", arr9_result)
```

10. Broadcasting Example
    Create a 4x4 array filled with 10s.

    Subtract a 1D array [0, 1, 2, 3] from each row using NumPy broadcasting.

```python
arr10 = np.full((4,4), 10)
arr10_result = arr10 - np.array([0,1,2,3])
print("\n10. Broadcasting Result:\n", arr10_result)
```
